Tic, Tac, Toe

```python
Theboard = {'top-L' : ' ',
            'top-M' : ' ',
            'top-R' : ' ',
            'mid-L' : ' ',
            'mid-M' : ' ',
            'mid-R' : ' ',
            'low-L' : ' ',
            'low-M' : ' ',
            'low-R' : ' ',}

def printBoard(board):
    print(board['top-L']+ '|' + board['top-M']+ '|' +board['top-R'])
    print('-+-+-')
    print(board['mid-L']+'|'+board['mid-M']+'|'+board['mid-R'])
    print('-+-+-')
    print(board['low-L']+'|'+board['low-M']+'|'+board['low-R'])
    
turn = 'x'
for i in range(8):
    printBoard(Theboard)
    
    print("It's " + turn + " enter your move\n")
    move = input()
    Theboard[move]=turn
    if turn == 'x':
        turn = 'o'
    else:
        turn = 'x'
printBoard(Theboard)
print("GAME OVER")
```

```python
Theboard = {'top-L' : ' ',
            'top-M' : ' ',
            'top-R' : ' ',
            'mid-L' : ' ',
            'mid-M' : ' ',
            'mid-R' : ' ',
            'low-L' : ' ',
            'low-M' : ' ',
            'low-R' : ' ',}

def printBoard(board):
    print(board['top-L']+ '|' + board['top-M']+ '|' +board['top-R'])
    print('-+-+-')
    print(board['mid-L']+'|'+board['mid-M']+'|'+board['mid-R'])
    print('-+-+-')
    print(board['low-L']+'|'+board['low-M']+'|'+board['low-R'])
    
turn = 'x'
for i in range(8):
    printBoard(Theboard)
    
    print("It's " + turn + " enter your move\n")
    move = input()
    Theboard[move]=turn
    if turn == 'x':
        turn = 'o'
    else:
        turn = 'x'
    if Theboard['top-L'] == Theboard['top-M'] and Theboard['top-M'] == Theboard['top-R'] and Theboard['top-R'] == Theboard['top-L']:
        print(turn + 'WON')
        break
    elif Theboard['mid-L'] == Theboard['mid-M'] and Theboard['mid-M'] == Theboard['mid-R'] and Theboard['mid-R'] == Theboard['mid-L']:
        print(turn + 'WON')
        break
    elif Theboard['low-L'] == Theboard['low-M'] and Theboard['low-M'] == Theboard['low-R'] and Theboard['low-R'] == Theboard['low-L']:
        print(turn + 'WON')
        break
    elif Theboard['top-L'] == Theboard['mid-M'] and Theboard['mid-M'] == Theboard['low-R'] and Theboard['low-R'] == Theboard['top-L']:
        print(turn + 'WON')
        break
    elif Theboard['top-R'] == Theboard['mid-M'] and Theboard['mid-M'] == Theboard['low-L'] and Theboard['low-L'] == Theboard['top-R']:
        print(turn + 'WON')
        break
    elif Theboard['top-L'] == Theboard['mid-L'] and Theboard['mid-L'] == Theboard['low-L'] and Theboard['low-L'] == Theboard['top-L']:
        print(turn + 'WON')
        break
    elif Theboard['top-M'] == Theboard['mid-M'] and Theboard['mid-M'] == Theboard['low-M'] and Theboard['low-M'] == Theboard['top-M']:
        print(turn + 'WON')
        break
    elif Theboard['top-R'] == Theboard['mid-R'] and Theboard['mid-R'] == Theboard['low-R'] and Theboard['low-R'] == Theboard['top-R']:
        print(turn + 'WON')
        break
printBoard(Theboard)
print("GAME OVER")
```
